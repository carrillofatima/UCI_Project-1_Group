```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import make_axes_locatable
```

```python
fp = 'Resources/cb_2018_us_state_500k/cb_2018_us_state_500k.shx'

map_df = gpd.read_file(fp, encoding="utf-8")
map_df = map_df.rename(index=str, columns={'STUSPS':'state_abbr'})

map_df.head()
```

```python
map_df.plot()
```

```python
df = pd.read_csv('Resources/state.csv', header=0)
df.head()
```

```python
merged = map_df.set_index('state_abbr').join(df.set_index('state_abbr'))
merged.head()
```

```python
poli_party = 'test'

vmin, vmax = 0, 23

fig, ax = plt.subplots(1, figsize=(10,6))

merged.plot(column=poli_party, cmaps['Diverging'] = [
            'PiYG', 'PRGn', 'BrBG', 'PuOr', 'RdGy', 'RdBu',
            'RdYlBu', 'RdYlGn', 'Spectral', 'coolwarm', 'bwr', 'seismic'], linewidth=0.8, ax=ax, edgecolor='0.8')
```
