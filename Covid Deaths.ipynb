```python
 %matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# variable to hold csv
file = 'Resources/us-states.csv'

# read the csv
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head()
```

```python
# pull just the dates, deaths and state
death_df = df[['date','state','deaths']]
death_df
```

```python
# convert date column to datetime in pandas 
death_df['date'] = pd.to_datetime(death_df['date'])
death_df['date']
```

```python
# create month column to group by
death_df['month'] = death_df['date'].dt.month_name()
death_df
```

```python
group_df = death_df.groupby(['state','month'])
cleaned_df = group_df['deaths'].count()
cleaned_df = pd.DataFrame(cleaned_df)
cleaned_df
```
