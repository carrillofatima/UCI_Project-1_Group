```python
 %matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# variable to hold csv
file = 'Resources/us-states.csv'

# read the csv
df = pd.read_csv(file, encoding="ISO-8859-1")
df.head()
```

```python
# pull just the dates, deaths and state
death_df = df[['state','date','deaths']]
death_df
```

```python
date_state_death = death_df[death_df["date"] == '2020-10-19']
date_state_death_df = pd.DataFrame(date_state_death)
#date_state_death_df
covid_death_oct_nineteenth = date_state_death_df.set_index("state")
covid_death_oct_ninteenth_df = pd.DataFrame(covid_death_oct_nineteenth)
covid_death_oct_ninteenth_df
```

```python
len(covid_death_oct_ninteenth_df)
```

```python
covid_death_oct_ninteenth_df
```

```python
narrow_total_state_deaths_df = covid_death_oct_ninteenth_df

final_total_death_state_df = narrow_total_state_deaths_df

final_total_death_state_df
```

```python
final_total_death_state_df.rename(columns = {'deaths':'Total Deaths from January 2020 to October 2020'}, inplace=True)
```

```python
# Changed column repesenting deaths to Total Deathsn from 01/2020 to 10/2020
final_total_death_state_df
```

```python
len(final_total_death_state_df)
```

```python
# Dropping date column, which was October 19th, 2020. So final df is the total amount of deaths as repoted by
#each state over the period of time begiining 1/21/2020 and endong on 10/19/2020

states_total_deaths = final_total_death_state_df.drop(['date'], axis=1)
states_total_deaths_df = pd.DataFrame(states_total_deaths)
states_total_deaths_df
```

```python
len(states_total_deaths_df)
```

```python
# The total number of COVID-19 deaths in the United States as of 10/19/2020, with the first data being reported 
# on 1/21/2020
states_total_deaths_df["Total Deaths from January 2020 to October 2020"].sum()
```
