```python
%matplotlib notebook
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# variable to hold csv
cody = 'Resources/death_df.csv'

# read the csv
cody_df = pd.read_csv(cody, encoding="ISO-8859-1")
cody_df.head()

# variable to hold csv
jackie = 'Resources/Jackie_pos_neg_total.csv'

# read the csv
jackie_df = pd.read_csv(jackie, encoding="ISO-8859-1")
jackie_df.head()

```

```python
cody_new = cody_df.rename(columns={'state':'States', 'month':'date'})

merged = jackie_df.merge(cody_new, on=['States', 'date'])
merged.groupby('States',sort='States')
merged
```

```python
merged.to_csv('Resources/state_covid_data.csv')
```
