```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import plotly.express as px
import geopandas as gpd

```

```python
# state lat and long df

state_coord_df = pd.read_csv('Resources/state_geocoord.csv', header=0)
state_coord_df = state_coord_df.rename(columns={'name':'States'})
state_coord_df.head()
```

```python
# poli and covid df

df = pd.read_csv('Resources/combined_df_ct.csv', header=0)
df.rename(columns={'Party Score':'party_score'})
df.head()
```

```python
# fip location df

fips = pd.read_csv('Resources/us-states.csv', dtype={'fips': str}, header=0)
fips = fips.rename(columns={'state':'States'})
fips.head()
```

```python
# merge poli and covid df with fips df

merged = df.merge(fips, on='States')
data = merged.merge(state_coord_df, on='States')
data.drop_duplicates()
new_data = data.rename(columns={'Party Score':'poli_score'})
new_data.head(30)
```

```python
new_data.to_csv('Resources/new_data.csv')
```

```python
# TEST* create choropleth map based on merged df and use fips as locations and party score as the color 

# set locations variable to lat and long
locations = new_data[['latitude','longitude']].astype(float)

fig = px.choropleth( new_data,locations=locations, color= new_data['poli_score'], 
                   color_continuous_scale='Viridis',
                   range_color = (0.0,23.0), 
                   scope= 'usa'
                   )

fig.update_layout(margin={r:'0',t:'0',l:'0',b:'0'})
fig.show()
```
